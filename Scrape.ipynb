```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.federalreserve.gov/monetarypolicy/fomccalendars.htm"
req = requests.get(url)
soup = BeautifulSoup(req.content, 'html.parser')
#print(soup.find_all(attrs = {'fomc-meeting__month'}))
query = soup.find_all('div', {'class':"panel panel-default"})
dates = []
months = []  

for i in range (6):
    query_months = query[i].find_all('div', {'class':"fomc-meeting__month"})
    query_dates = query[i].find_all('div', {'class':"fomc-meeting__date"})


    for date in query_dates:
        dates.append(date.text)


    for month in query_months:
        months.append(month.text)
    
display(dates)
print(months)

#for date in dates_2022:

# query_2021_months = query[1].find_all('div', {'class':"fomc-meeting__month"})

# months_2021 = []

# for month in query_2021_months:
#     months_2021.append(month.text)
# display(months_2021)
```

```python
years = [" 2022"," 2021", " 2020", " 2019", " 2018", " 2017"]
year = 0
counter = 0
  

while year <6:
    while months[counter] != "December":
        months[counter] = months[counter] + years[year]
        counter = counter + 1
    months[counter] = months[counter] + years[year]
    year = year + 1
    counter = counter + 1

months
        
```

```python
def date_cleaner (dates_year):
    split_dates_2022 = []

    for date in dates_year:
        split_dates_2022.append(re.split(r"\D",date))
    #print(split_dates_2022)
    
    ssplit_dates_2022 =[]
    
    for date in split_dates_2022:
        ssplit_dates_2022.append(date[0].split("*",1))
    #print(ssplit_dates_2022)
    
    # sssplit_dates_2022 = []    
    # for date in ssplit_dates_2022:
    #     ssplit_dates_2022.append(date.split(" ",1))
    # print(sssplit_dates_2022)

    flat_list = []
    for sublist in ssplit_dates_2022:
        for item in sublist:
            flat_list.append(item)

    flat_list = list(map(int, flat_list))

    
    fflat_list = []        
        
    for item in flat_list:
        if item == 31:
            item = 1
            fflat_list.append(item)
        else:
            fflat_list.append(item + 1)
        
    #Undo the +1 for certain indexes that are in a different format on the source URL
    fflat_list[17] = fflat_list[17]-1
    fflat_list[18] = fflat_list[18]-1
    fflat_list[20] = fflat_list[20]-1
    fflat_list[21] = fflat_list[21]-1
    fflat_list[22] = fflat_list[22]+30
    fflat_list[26] = fflat_list[26]-1
    fflat_list[32] = fflat_list[32]-30
    fflat_list[36] = fflat_list[36]-1
    
     
    return(fflat_list)
```

```python
dates = date_cleaner(dates)
dates
```
